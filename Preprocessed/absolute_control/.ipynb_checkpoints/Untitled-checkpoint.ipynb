```python
def get_cat(df, value):
    if value == 'CCI_flags':
        full_categories = list(range(17))
    elif value == 'race_ethnicity':
        full_categories = ['NHB', 'NHW', 'Hispanic', 'Other', 'Unknown']
    elif value == 'insurance_types':
        full_categories = ['Medicaid', 'EHR']
    elif value == 'smoking':
        full_categories = [4, 3, 2, 1]
    else:
        full_categories = [0, 1]
    count = pd.DataFrame(df[value].value_counts().reindex(full_categories, fill_value=0))
    perc = pd.DataFrame(round((df[value].value_counts().reindex(full_categories, fill_value=0)/df.shape[0])*100, 2))
    combine = pd.concat([count,perc], axis = 1)
    combine['combine'] = combine.apply(
        lambda row: f"{str(row.iloc[0])} ({row.iloc[1]}%)", axis=1
    )
    combine = pd.DataFrame(combine['combine'])
    combine = combine.reset_index()
    name = pd.DataFrame([value, np.nan]).T
    name.columns = combine.columns
    combine = pd.concat([name, combine], ignore_index=True)
    combine.columns = [0,1]
    return combine
def get_value(df, value):
    mean = round(df[value].mean(), 2)
    std = round(df[value].std(),2)
    temp = pd.DataFrame(['Mean (SD)', f'{mean} ({std})']).T
    temp1 = pd.DataFrame([str(value), np.nan]).T
    return pd.concat([temp1, temp])
def get_sta(df, outcome):
    whole = pd.DataFrame()
    count = df.shape[0]
    count = f'(N = {count})'
    head = pd.concat([pd.DataFrame([np.nan, str(outcome)]).T, pd.DataFrame([np.nan, count]).T])
    whole = pd.concat([whole, head])
    for i in value_cols:
        temp = get_value(df, i)
        missing_rate = round((df[i].isna().sum()/df.shape[0])*100,2)
        temp = pd.concat([temp, pd.DataFrame(['Missing Rate', str(df[i].isna().sum()) + f' ({missing_rate}%)']).T])
        whole = pd.concat([whole, temp])
    for i in cat_cols:
        temp = get_cat(df, i)
        whole = pd.concat([whole, temp])
    return whole.reset_index(drop=True)
def get_outcome(case):
    hsr = case[case.case_HSR_flag == 1]
    cp1 = case[case.case_CP1_flag == 1]
    cp2 = case[case.case_CP2_flag == 1]
    cp1_cp2_hsr = case[case['cp1+cp2+hsr'] == 1]
    whole = pd.DataFrame()
    temp = get_sta(hsr, 'HSR')
    whole = pd.concat([whole, temp], axis = 1)
    temp = get_sta(cp1, 'CP1')
    whole = pd.concat([whole, temp], axis = 1)
    temp = get_sta(cp2, 'CP2')
    whole = pd.concat([whole, temp], axis = 1)
    temp = get_sta(cp1_cp2_hsr, 'cp1 & cp2 & hsr')
    whole = pd.concat([whole, temp], axis = 1)
    return whole
```
